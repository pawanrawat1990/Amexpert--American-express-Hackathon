```python
import numpy as np
import pandas as pd
```

```python
import pandas as pd
import zipfile
#### Making a connection with zip file ###################################
zf = zipfile.ZipFile("D:/Python/Amexpert/train_AUpWtIz.zip")
```

```python
#### Importing csv files from zip file ############################
train=pd.read_csv(zf.open("train.csv"))
campaign_data=pd.read_csv(zf.open("campaign_data.csv"))
coupon_item_mapping=pd.read_csv(zf.open("coupon_item_mapping.csv"))
customer_demographics=pd.read_csv(zf.open("customer_demographics.csv"))
customer_transaction_data=pd.read_csv(zf.open("customer_transaction_data.csv"))
item_data=pd.read_csv(zf.open("item_data.csv"))
```

```python
#### Importing test and submission file ###################################
test=pd.read_csv("D:/Python/Amexpert/test_QyjYwdj.csv")
Submission=pd.read_csv("D:/Python/Amexpert/sample_submission_Byiv0dS.csv")
```

```python
test["redemption_status"]=0
train["Temp"]= "train"
test["Temp"]="test"
```

```python
combi=pd.concat([train,test], axis=0,ignore_index=True)
```

combi

```python
combi
```

```python
Coupon_Item_Data= pd.merge(coupon_item_mapping,item_data,on=["item_id"],how="left")
```

```python
Coupon_Item_Transac=pd.merge(Coupon_Item_Data,customer_transaction_data,on=["item_id"],how="left")
```

```python
Grouping_Item_Transac= Coupon_Item_Transac.groupby(['coupon_id','customer_id']).aggregate({'quantity':'sum','selling_price':'sum','other_discount':'sum','coupon_discount':'sum','brand_type':'max','category':'max'})
```

```python
combi_transac=pd.merge(combi,Grouping_Item_Transac,on=["coupon_id"],how='left')
```

```python
data=combi_transac.groupby(["id"]).aggregate({'quantity':'sum','selling_price':'sum','other_discount':'sum','coupon_discount':'sum','brand_type':'max','category':'max'})
```

```python
data_1=pd.merge(combi,data,on=["id"],how='left')
```

```python
final=pd.merge(data_1,campaign_data,on=["campaign_id"],how='left')
```

```python
import datetime
```

```python
final["start_date"]= pd.to_datetime(final["start_date"],format="%d/%m/%y")
```

```python
final["end_date"]= pd.to_datetime(final["end_date"],format="%d/%m/%y")
```

```python
final["diff"]=final["end_date"]-final["start_date"]
```

```python
final["campaign_duration_days"]=final["diff"].dt.days
```

```python
final
```

```python
num_unique_items=coupon_item_mapping.groupby(["coupon_id"])["item_id"].nunique()
```

```python
final_1=pd.merge(final,num_unique_items,on=["coupon_id"],how="left")
```

```python
final_1.rename(columns={"item_id": "num_unique_items"},inplace=True)
```

```python
final_all=pd.merge(final_1,customer_demographics,on=["customer_id"],how="left")
```

```python
final_all.apply(lambda x: sum(x.isnull()))
```

```python
final_all["family_size"]=pd.to_numeric(final_all["family_size"].str.replace("[^0-9.]","",regex=True))
```

```python
######## Assuming family size is 1 for missing values ###
final_all["family_size"].fillna(1, inplace=True)
```

```python
####### Filling rest of variables missing values with NA to easily filter out rows #################
final_all.fillna("NA",inplace=True)
```

```python
final_all.loc[(final_all["family_size"]==1) & (final_all["marital_status"]== "NA"),"marital_status"]= "Single"
```

```python
##### Assuming Rest are married as family size is more than 1.
final_all.loc[final_all["marital_status"]=="NA","marital_status"]="Married"
```

```python
final_all.loc[(final_all["family_size"]<=2) & (final_all["no_of_children"]== "NA"),"no_of_children"]= "0"
```

```python
final_all.loc[(final_all["marital_status"]=="Single") & (final_all["age_range"]== "NA"),"age_range"]= "18-25"
```

```python
# sns.countplot(x="age_range", data=final_all,hue="marital_status",palette="Greens_d")
# # adding another categorical variable
```

```python
final_all["no_of_children"]=pd.to_numeric(final_all["no_of_children"].str.replace("[^0-9]","",regex=True))
```

```python
final_all.loc[final_all["income_bracket"]=="NA","income_bracket"]=0
```

```python
final_all["income_bracket"]=pd.to_numeric(final_all["income_bracket"])
```

```python
final_all.groupby(["age_range"])["income_bracket"].mean()
```

```python
#### Taking mean as 4 , this is most common #######
final_all.loc[final_all["income_bracket"]==0,"income_bracket"]=4
```

```python
final_all["age_range"]=final_all["age_range"].str.replace("+","-90",regex=True)
```

```python
final_all['age_1'], final_all['age_2'] = zip(*final_all['age_range'].apply(lambda x: x.split('-', 1)))
```

```python
final_all["age_1"]=pd.to_numeric(final_all["age_1"])
final_all["age_2"]=pd.to_numeric(final_all["age_2"])
```

```python
final_all["age"]=round((final_all["age_1"]+final_all["age_2"])/2)
```

```python
dummy=pd.get_dummies(final_all[["brand_type","category","campaign_type","marital_status"]],drop_first=False)
mydata=pd.concat([final_all,dummy],axis=1)
```

```python
mydata.columns
```

```python
mydata.drop(["brand_type","category","campaign_type","start_date","end_date","coupon_id","customer_id","campaign_id","id","diff","marital_status","rented","age_range","age_1","age_2"],axis=1,inplace=True)
```

```python
mydata["other_discount"]=abs(mydata["other_discount"])
```

```python
mydata["coupon_discount"]=abs(mydata["coupon_discount"])
```

```python
mydata.describe(percentiles=[0.01,0.75,0.90,0.95,0.99]).T
```

```python
num_col= mydata.select_dtypes(exclude="object")
numeric_col=num_col[["quantity","selling_price","other_discount","coupon_discount","num_unique_items"]]
for i in numeric_col.dtypes.index:
    mydata.loc[mydata[i]>mydata[i].quantile(0.95),i]=mydata[i].quantile(0.95)
```

```python
training=mydata.loc[mydata["Temp"]=="train"]
testing=mydata.loc[mydata["Temp"]=="test"]
```

```python
training.drop(["Temp"],axis=1,inplace=True)
testing.drop(["Temp","redemption_status"],axis=1,inplace=True)
```

```python
x_train=training.drop(["redemption_status"],1)
y_train=training["redemption_status"]
```

```python
from sklearn.model_selection import KFold, train_test_split
```

```python
np.random.seed(324)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.30,random_state=43)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```
